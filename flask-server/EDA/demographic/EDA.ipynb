```python
import pandas as pd

df = pd.read_csv('../../dataset/cleaned/demographic.csv', header=1, index_col=0)
```

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
plt.rcParams['font.family']='Tahoma'
plt.rcParams['font.size']=13

def extract_indexes(df):
    province_names = df['ชื่อจังหวัด'].unique()
    registry_names = df['ชื่อสำนักทะเบียน'].unique()
    district_names = df['ชื่อตำบล'].unique()
    print(province_names, len(province_names))
    print(registry_names, len(registry_names))
    print(district_names, len(district_names))
    return province_names, registry_names, district_names

def extract_age_brackets(df):
    df['1519'] = df['ชาย 15 ปี'] + df['ชาย 16 ปี'] + df['ชาย 17 ปี'] + df['ชาย 18 ปี'] + df['ชาย 19 ปี']
    + df['หญิง 15 ปี'] + df['หญิง 16 ปี'] + df['หญิง 17 ปี'] + df['หญิง 18 ปี'] + df['หญิง 19 ปี']
    df['2024'] = df['ชาย 20 ปี'] + df['ชาย 21 ปี'] + df['ชาย 22 ปี'] + df['ชาย 23 ปี'] + df['ชาย 24 ปี']
    + df['หญิง 20 ปี'] + df['หญิง 21 ปี'] + df['หญิง 22 ปี'] + df['หญิง 23 ปี'] + df['หญิง 24 ปี']
    df['2529'] = df['ชาย 25 ปี'] + df['ชาย 26 ปี'] + df['ชาย 27 ปี'] + df['ชาย 28 ปี'] + df['ชาย 29 ปี']
    + df['หญิง 25 ปี'] + df['หญิง 26 ปี'] + df['หญิง 27 ปี'] + df['หญิง 28 ปี'] + df['หญิง 29 ปี']
    df['3034'] = df['ชาย 30 ปี'] + df['ชาย 31 ปี'] + df['ชาย 32 ปี'] + df['ชาย 33 ปี'] + df['ชาย 34 ปี']
    + df['หญิง 30 ปี'] + df['หญิง 31 ปี'] + df['หญิง 32 ปี'] + df['หญิง 33 ปี'] + df['หญิง 34 ปี']
    df['3539'] = df['ชาย 35 ปี'] + df['ชาย 36 ปี'] + df['ชาย 37 ปี'] + df['ชาย 38 ปี'] + df['ชาย 39 ปี']
    + df['หญิง 35 ปี'] + df['หญิง 36 ปี'] + df['หญิง 37 ปี'] + df['หญิง 38 ปี'] + df['หญิง 39 ปี']
    df['4044'] = df['ชาย 40 ปี'] + df['ชาย 41 ปี'] + df['ชาย 42 ปี'] + df['ชาย 43 ปี'] + df['ชาย 44 ปี']
    + df['หญิง 40 ปี'] + df['หญิง 41 ปี'] + df['หญิง 42 ปี'] + df['หญิง 43 ปี'] + df['หญิง 44 ปี']
    df['4549'] = df['ชาย 45 ปี'] + df['ชาย 46 ปี'] + df['ชาย 47 ปี'] + df['ชาย 48 ปี'] + df['ชาย 49 ปี']
    + df['หญิง 45 ปี'] + df['หญิง 46 ปี'] + df['หญิง 47 ปี'] + df['หญิง 48 ปี'] + df['หญิง 49 ปี']
    df['5054'] = df['ชาย 50 ปี'] + df['ชาย 51 ปี'] + df['ชาย 52 ปี'] + df['ชาย 53 ปี'] + df['ชาย 54 ปี']
    + df['หญิง 50 ปี'] + df['หญิง 51 ปี'] + df['หญิง 52 ปี'] + df['หญิง 53 ปี'] + df['หญิง 54 ปี']
    df['5559'] = df['ชาย 55 ปี'] + df['ชาย 56 ปี'] + df['ชาย 57 ปี'] + df['ชาย 58 ปี'] + df['ชาย 59 ปี']
    + df['หญิง 55 ปี'] + df['หญิง 56 ปี'] + df['หญิง 57 ปี'] + df['หญิง 58 ปี'] + df['หญิง 59 ปี']
    return df

def extract_registry_stats(df, registry_names):
    registry_df = df.groupby(['ชื่อสำนักทะเบียน', 'ปีเดือน']).sum()[['ผลรวมประชากรทั้งหมด', 'ผลรวมประชากรที่อยู่ระหว่างการย้าย']]
    registry_df = registry_df.apply(pd.to_numeric, errors='coerce')
    registry_df['outflow_percentage'] = registry_df['ผลรวมประชากรที่อยู่ระหว่างการย้าย'] * 100 / registry_df['ผลรวมประชากรทั้งหมด']

    with PdfPages('output/bkk_population_evolution_by_registry.pdf') as pdf:
        for registry in registry_names:
            plot_df = registry_df.reset_index()
            plot_df = plot_df[plot_df['ชื่อสำนักทะเบียน'] == registry]
            plot_df.plot(x="ปีเดือน", y=["ผลรวมประชากรทั้งหมด"], title=registry)
            pdf.savefig()
            plt.close()

def extract_district_stats(df, district_names):
    district_df = df.groupby(['ชื่อตำบล', 'ปีเดือน']).sum()[['ผลรวมประชากรทั้งหมด', 'ผลรวมประชากรที่อยู่ระหว่างการย้าย']]
    district_df = district_df.apply(pd.to_numeric, errors='coerce')
    district_df['outflow_percentage'] = district_df['ผลรวมประชากรที่อยู่ระหว่างการย้าย'] * 100 / district_df['ผลรวมประชากรทั้งหมด']

    with PdfPages('output/bkk_population_evolution_by_district.pdf') as pdf:
        for district in district_names:
            plot_df = district_df.reset_index()
            plot_df = plot_df[plot_df['ชื่อตำบล'] == district]
            plot_df.plot(x="ปีเดือน", y=["ผลรวมประชากรทั้งหมด"], title=district)
            pdf.savefig()
            plt.close()

def extract_registry_age_stats(df, registry_names):
    registry_df_age = df.groupby(['ชื่อสำนักทะเบียน', 'ปีเดือน']).sum()
    registry_df_age = extract_age_brackets(registry_df_age)

    with PdfPages('output/bkk_population_evolution_by_registry_age.pdf') as pdf:
        for registry in registry_names:
            plot_df = registry_df_age.reset_index()
            plot_df = plot_df[plot_df['ชื่อสำนักทะเบียน'] == registry]
            plot_df.plot(x="ปีเดือน", y=["1519", "2024", "2529", "3034", "3539", "4044", "4549", "5054", "5559"], title=registry)
            pdf.savefig()
            plt.close()

def extract_district_age_stats(df, district_names):
    district_df_age = df.groupby(['ชื่อตำบล', 'ปีเดือน']).sum()
    district_df_age = extract_age_brackets(district_df_age)

    with PdfPages('output/bkk_population_evolution_by_district_age.pdf') as pdf:
        for district in district_names:
            plot_df = district_df_age.reset_index()
            plot_df = plot_df[plot_df['ชื่อตำบล'] == district]
            plot_df.plot(x="ปีเดือน", y=["1519", "2024", "2529", "3034", "3539", "4044", "4549", "5054", "5559"], title=district)
            pdf.savefig()
            plt.close()

bangkok_df = df[df['ชื่อจังหวัด']=='กรุงเทพมหานคร']
province_names, registry_names, district_names = extract_indexes(bangkok_df)
extract_registry_stats(bangkok_df, registry_names)
extract_district_stats(bangkok_df, district_names)
extract_registry_age_stats(bangkok_df, registry_names)
extract_district_age_stats(bangkok_df, district_names)
```
